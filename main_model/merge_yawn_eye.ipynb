```python
import os
import shutil  # For file deletion

# image_path_v5 = os.path.join(yawnv5, 'train', 'images')
# label_path_v5 = os.path.join(yawnv5, 'train', 'labels')

def delete_empty_files(image_path, label_path):
    for file in os.listdir(label_path):
        file_path = os.path.join(label_path, file)

        # Read file content
        with open(file_path, 'r') as f:
            content = f.read().strip()

        # Check if file is empty or only contains whitespace
        if not content:
            # Find corresponding image file
            image_file = os.path.splitext(file)[0] + '.jpg'  # Assuming jpg extension
            # Also check for other common extensions if needed
            if not os.path.exists(os.path.join(image_path, image_file)):
                image_file = os.path.splitext(file)[0] + '.png'
                if not os.path.exists(os.path.join(image_path, image_file)):
                    image_file = os.path.splitext(file)[0] + '.jpeg'

            image_file_path = os.path.join(image_path, image_file)

            # Delete both files if image exists
            os.remove(file_path)
            print(f"Deleted empty label file: {file}")

            if os.path.exists(image_file_path):
                os.remove(image_file_path)
                print(f"Deleted corresponding image: {image_file}")
            else:
                print(f"No corresponding image found for: {file}")
        else:
            # Proceed with modification for non-empty files
            mod_content = '2' + content[1:]

            with open(file_path, 'w') as f:
                f.write(mod_content)

            # Confirm proper change
            with open(file_path, 'r') as f:
                updated_content = f.read()
                print(f"Updated: {file} - {updated_content[:50]}...")
```

```python

image_path_v5 = os.path.join(yawnv5, 'train', 'images')
label_path_v5 = os.path.join(yawnv5, 'valid', 'labels')

delete_empty_files(image_path_v5, label_path_v5)
```

```python
def delete_empty_filesv3(folder_path):
    """
    Delete empty label files and their corresponding images when they're in the same folder.
    Also changes class ID from 0 to 2 for non-empty label files.
    """
    # Get all label files (txt files)
    label_files = [f for f in os.listdir(folder_path) if f.endswith('.txt')]
    
    for file in label_files:
        file_path = os.path.join(folder_path, file)
        
        # Read file content
        with open(file_path, 'r') as f:
            content = f.read().strip()
            
        # Check if file is empty or only contains whitespace
        if not content:
            # Delete the empty label file
            os.remove(file_path)
            print(f"Deleted empty label file: {file}")
            
            # Find corresponding image file
            base_name = os.path.splitext(file)[0]
            image_found = False
            
            for ext in ['.jpg', '.png', '.jpeg']:
                image_file = base_name + ext
                image_file_path = os.path.join(folder_path, image_file)
                
                if os.path.exists(image_file_path):
                    os.remove(image_file_path)
                    print(f"Deleted corresponding image: {image_file}")
                    image_found = True
                    break
                    
            if not image_found:
                print(f"No corresponding image found for: {file}")
        else:
            # Proceed with modification for non-empty files
            mod_content = '2' + content[1:]
            
            with open(file_path, 'w') as f:
                f.write(mod_content)
                
            # Confirm proper change
            with open(file_path, 'r') as f:
                updated_content = f.read()
                print(f"Updated: {file} - {updated_content[:50]}...")
```

```python
image_path_v3 = os.path.join(yawnv3, 'valid')

delete_empty_filesv3(image_path_v3)
```

```python
combined_data_v5 = Path('/Users/seunghyunlee/Desktop/aps360_project/dataset/combined_v5')
from pathlib import Path
import os

def check_image_label_correspondence(dataset_path, delete_unpaired_images=False):
    """
    Check if each image has a corresponding label file in train, valid, and test splits.
    Optionally deletes images that don't have label files.
    
    Args:
        dataset_path: Path to the dataset root folder
        delete_unpaired_images: If True, will delete images without labels
    """
    dataset_path = Path(dataset_path)
    splits = ['train', 'valid', 'test']
    total_deleted = 0
    
    for split in splits:
        print(f"\n--- Checking {split} split ---")
        
        # Define paths
        images_dir = dataset_path / split / 'images'
        labels_dir = dataset_path / split / 'labels'
        
        if not images_dir.exists() or not labels_dir.exists():
            print(f"Missing directory: {split} split has issues. Please check the folder structure.")
            continue
        
        # Get lists of files
        image_files = [f for f in images_dir.glob('*.jpg')]
        image_stems = [f.stem for f in image_files]
        label_files = [f.stem for f in labels_dir.glob('*.txt')]
        
        # Check image files without label files
        images_without_labels_stems = set(image_stems) - set(label_files)
        images_without_labels = [images_dir / f"{stem}.jpg" for stem in images_without_labels_stems]
        
        if images_without_labels:
            print(f"Found {len(images_without_labels)} images without labels:")
            
            # Delete unpaired images if requested
            if delete_unpaired_images:
                for img_path in images_without_labels:
                    print(f"  - Deleting {img_path.name}")
                    img_path.unlink()
                    total_deleted += 1
            else:
                # Just print the first 10 if not deleting
                for img in sorted([img.stem for img in images_without_labels])[:10]:
                    print(f"  - {img}.jpg")
                if len(images_without_labels) > 10:
                    print(f"  ... and {len(images_without_labels) - 10} more")
        else:
            print("All images have corresponding label files.")
        
        # Check label files without image files (optional)
        labels_without_images = set(label_files) - set(image_stems)
        if labels_without_images:
            print(f"Found {len(labels_without_images)} labels without images:")
            for lbl in sorted(list(labels_without_images))[:10]:  # Show first 10
                print(f"  - {lbl}.txt")
            if len(labels_without_images) > 10:
                print(f"  ... and {len(labels_without_images) - 10} more")
        else:
            print("All label files have corresponding images.")
        
        # Print summary
        print(f"Summary for {split}:")
        print(f"  - Total images: {len(image_stems)}")
        print(f"  - Total labels: {len(label_files)}")
        print(f"  - Images without labels: {len(images_without_labels)}")
        print(f"  - Labels without images: {len(labels_without_images)}")
    
    if delete_unpaired_images:
        print(f"\nTotal images deleted: {total_deleted}")

# Run the check with deletion
# Set delete_unpaired_images=True to actually delete the images without labels
check_image_label_correspondence(combined_data_v5, delete_unpaired_images=True)
```

```python

from pathlib import Path
import os

def check_image_label_correspondence_same_folder(folder_path, delete_unpaired_images=False):
    """
    Check if each image has a corresponding label file when they're in the same folder.
    Optionally deletes images that don't have label files.
    
    Args:
        folder_path: Path to the folder containing both images and labels
        delete_unpaired_images: If True, will delete images without labels
    """
    folder_path = Path(folder_path)
    total_deleted = 0
    
    if not folder_path.exists():
        print(f"Folder not found: {folder_path}")
        return
    
    print(f"\n--- Checking {folder_path} ---")
    
    # Get lists of files
    image_files = list(folder_path.glob('*.jpg'))
    image_stems = [f.stem for f in image_files]
    label_files = list(folder_path.glob('*.txt'))
    label_stems = [f.stem for f in label_files]
    
    # Check image files without label files
    images_without_labels_stems = set(image_stems) - set(label_stems)
    images_without_labels = [folder_path / f"{stem}.jpg" for stem in images_without_labels_stems]
    
    if images_without_labels:
        print(f"Found {len(images_without_labels)} images without labels:")
        
        # Delete unpaired images if requested
        if delete_unpaired_images:
            for img_path in images_without_labels:
                print(f"  - Deleting {img_path.name}")
                img_path.unlink()
                total_deleted += 1
        else:
            # Just print the first 10 if not deleting
            for img in sorted([img.stem for img in images_without_labels])[:10]:
                print(f"  - {img}.jpg")
            if len(images_without_labels) > 10:
                print(f"  ... and {len(images_without_labels) - 10} more")
    else:
        print("All images have corresponding label files.")
    
    # Check label files without image files (optional)
    labels_without_images = set(label_stems) - set(image_stems)
    if labels_without_images:
        print(f"Found {len(labels_without_images)} labels without images:")
        for lbl in sorted(list(labels_without_images))[:10]:  # Show first 10
            print(f"  - {lbl}.txt")
        if len(labels_without_images) > 10:
            print(f"  ... and {len(labels_without_images) - 10} more")
    else:
        print("All label files have corresponding images.")
    
    # Print summary
    print(f"Summary:")
    print(f"  - Total images: {len(image_files)}")
    print(f"  - Total labels: {len(label_files)}")
    print(f"  - Images without labels: {len(images_without_labels)}")
    print(f"  - Labels without images: {len(labels_without_images)}")
    
    if delete_unpaired_images:
        print(f"\nTotal images deleted: {total_deleted}")
```
