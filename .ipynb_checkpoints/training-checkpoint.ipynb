```python
import cv2
from ultralytics import YOLO

# Load the model - ensure you're using the best weights from your training
model = YOLO("runs/detect/train182/weights/yolo_zach.pt")  # Use your trained model path

cap = cv2.VideoCapture(0)
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break
    
    # Resize frame to match the training image size (160x96)
    resized_frame = cv2.resize(frame, (160, 96))
    
    # Make detections with lower confidence threshold
    results = model(resized_frame, conf=0.25)  # Lower confidence threshold
    
    # Access the first result
    result = results[0]
    
    # Check if there are any detections
    if len(result.boxes) > 0:
        # Draw the detections on a copy of the original frame
        # We'll scale the bounding boxes back to the original frame size
        rendered_frame = frame.copy()
        h, w, _ = frame.shape
        scale_x = w / 160
        scale_y = h / 96
        
        for box in result.boxes:
            # Get box coordinates, class and confidence
            x1, y1, x2, y2 = map(int, box.xyxy[0])
            
            # Scale coordinates to original frame size
            x1, x2 = int(x1 * scale_x), int(x2 * scale_x)
            y1, y2 = int(y1 * scale_y), int(y2 * scale_y)
            
            cls = int(box.cls[0])
            conf = float(box.conf[0])
            class_name = result.names[cls]
            
            # Draw rectangle and label
            color = (0, 255, 0) if class_name == 'opened' else (0, 0, 255)
            cv2.rectangle(rendered_frame, (x1, y1), (x2, y2), color, 2)
            label = f"{class_name} {conf:.2f}"
            cv2.putText(rendered_frame, label, (x1, y1 - 10), 
                        cv2.FONT_HERSHEY_SIMPLEX, 0.9, color, 2)
    else:
        rendered_frame = frame  # If no boxes, show the original frame
    
    # Display the image
    cv2.imshow('YOLO', rendered_frame)
    
    if cv2.waitKey(10) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
cap = cv2.VideoCapture(0)
while cap.isOpened():
    ret, frame = cap.read()
    
    # Make detections 
    results = model(frame)
    
    # Access the first result (since it's a list of results)
    result = results[0]

   # Filter the results to keep detections of 'closed' (class 0) and 'opened' (class 1)
    class_names = result.names  # Access class names from the model
    filtered_boxes = []

    # Iterate through detected boxes and filter by class name
    for box, cls in zip(result.boxes, result.boxes.cls):
        class_name = class_names[int(cls)]  # Get the class name for the detected box
        if class_name in ['closed', 'opened']:  # Filter for 'closed' and 'opened'
            filtered_boxes.append(box)

    # If there are any filtered boxes, plot them
    if filtered_boxes:
        # You can use result.plot() to render these filtered results on the image
        result.boxes = filtered_boxes  # Update the boxes to only include filtered ones
        rendered_frame = result.plot()  # Render the frame with the filtered detections
    else:
        rendered_frame = frame  # If no boxes, just show the original frame
    
    # Display the image
    cv2.imshow('YOLO', rendered_frame)
    
    if cv2.waitKey(10) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
